```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
months=pd.date_range('2018-01',periods=12,freq='M')
max_temp=[39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25]
min_temp=[21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18]
df=pd.DataFrame(max_temp,index=months)
df.columns=['max_temp']
df['min_temp']=min_temp
df.plot()
```
